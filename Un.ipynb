### Recommender System Using TFIDF

```python
#TF IDF Vector 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Define a TF-IDF Vectorizer Object. Remove all english stopwords
tfidf = TfidfVectorizer(stop_words='english')
```

```python
# Since we can't pass a list as a variable we convert it into a string
cred_df['cast'] = cred_df.apply(lambda x: ', '.join(x['cast']),axis=1)
```

```python
# Since we can't pass a list as a variable we convert it into a string
cred_df['crew'] = cred_df.apply(lambda x: ', '.join(x['crew']),axis=1)
```

```python
# Since we can't pass a list as a variable we convert it into a string
key_df['keywords'] = key_df.apply(lambda x: ', '.join(x['keywords']),axis=1)
```

```python
# creating the combined data frame
com_df = pd.merge(key_df,cred_df,on='id')
com_df.head()
```

```python
#Construct the required TF-IDF matrix by applying the fit_transform method on the overview feature
tfidf_cast = tfidf.fit_transform(com_df['cast']) #for cast
tfidf_crew = tfidf.fit_transform(com_df['crew']) #for crew
tfidf_keys = tfidf.fit_transform(com_df['keywords']) #for keywords
```

```python
tfidf_cast.shape
```

```python
# Import linear_kernel to compute the dot product
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix for cast
cosine_cast = linear_kernel(tfidf_cast, tfidf_cast)
```

```python
# Compute the cosine similarity matrix for crew
cosine_crew = linear_kernel(tfidf_crew, tfidf_crew)
```

```python
# Compute the cosine similarity matrix for keywords
cosine_keys = linear_kernel(tfidf_keys, tfidf_keys)
```

```python
# with id of the movie we are sorting based on similarity score for crew
[(cosine_crew[200][i], com_df['id'][i]) for i in cosine_crew[200].argsort()[:-20:-1]]
# reverse mapping of movieids to get indices
indices = pd.Series(com_df.index, index=com_df['id']).drop_duplicates()
indices
```

```python
# with id of the movie we are sorting based on similarity score for keywords
[(cosine_keys[200][i], com_df['id'][i]) for i in cosine_keys[200].argsort()[:-20:-1]]
# reverse mapping of movieids to get indices
indices = pd.Series(com_df.index, index=com_df['id']).drop_duplicates()
indices
```

```python
# Function that takes in cast as input and gives recommendations 
def content_recommender(cast, cosine_sim=cosine_cast, df=com_df, indices=indices):

    idx = indices[cast]

    # Get the pairwsie similarity scores of all movies with that cast member
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the scores of the 10 most similar movies. Ignore the first movie.
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['cast'].iloc[movie_indices]

```
